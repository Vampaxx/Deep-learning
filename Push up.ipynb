```python
import cv2 
import mediapipe as mp
import numpy as np 
from matplotlib import pyplot as plt 

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
##
cap = cv2.VideoCapture('http://192.168.20.4:8080/video')
with mp_pose.Pose(min_detection_confidence = 0.5, min_tracking_confidence = 0.5) as pose :
    while cap.isOpened():
        ret ,frame = cap.read()
        # recolor the frame 
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False ## saving the memory 
        
        # make detection 
        results = pose.process(image)
        
        # recoloring the image back to BGR
        image.flags.writeable = True 
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)## open cv need BGR format
        
        # Render the detection
        mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(120,255,12),thickness=2, circle_radius=4),## points
                                 mp_drawing.DrawingSpec(color=(245,66,230),thickness=4, circle_radius=4),## connection
                                 )
        
        ## make pixel format
        resized_image = cv2.resize(image,(640,480))
        cv2.imshow('mediapipe feed',cv2.flip(resized_image,1))   # flip the image horizondaly for a selfi-view display          
        



        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()


```

```python
def calculate_angle(a,b,c):
    
    a = np.array(a) ## joints fisrt 
    b = np.array(b) ## joints mid
    c = np.array(c) ## joints last
    
    radian = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle  = np.abs(radian * 180.0 / np.pi)
                    
    if angle > 180.0 :
        return 360.0-angle
    return angle
```
